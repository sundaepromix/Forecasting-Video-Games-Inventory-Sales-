```python
# ADF Test to check for stationarity
adf_result = adfuller(df['Monthly Sales'])
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
```

```python
def check_stationarity(timeseries):
    result = adfuller (timeseries)
    print('Augmented Dickey-Fuller Test Results:')
    print('ADF Statistics:', result[0])
    print('p-value:', result[1])
    print('Critical values:')
    for key, value in result[4].items():
        print(f'\t{key}): {value}')
    return result[1] < 0.05

is_stationary = check_stationarity (df['Monthly Sales'])
print(f"\nTime Series is {'stationary' if is_stationary else 'non-stattionary'}")
```

## ARIMA Model

```python
import pandas as pd
import numpy as np
import pmdarima as pm
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 1: Load and Prepare the Data
# Replace 'df' with your actual DataFrame and ensure 'Monthly Sales' is the target column
#df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-01')
#df.set_index('Date', inplace=True)

# Step 2: Fit the auto_arima model to identify optimal ARIMA order
auto_model = pm.auto_arima(df['Monthly Sales'], 
                            seasonal=True,   
                            m=12,            
                            stepwise=True,   
                            trace=True,      
                            suppress_warnings=True,  
                            seasonal_test='ch',  
                            max_p=5,         
                            max_q=5,         
                            max_P=2,         
                            max_Q=2,         
                            d=0,             
                            D=0,             
                            max_order=10)    

# Extract optimal ARIMA orders
order = auto_model.order
seasonal_order = auto_model.seasonal_order
print("\nOptimal ARIMA order:", order)
print("Optimal Seasonal Order:", seasonal_order)
```

```python

# Step 3: Fit ARIMA model with the extracted order
model = ARIMA(df['Monthly Sales'], order=order, seasonal_order=seasonal_order)
results = model.fit()

# Step 4: Print ARIMA model summary
print("\nModel Summary:")
print(results.summary())
```

```python

# Step 5: Calculate and display model performance metrics
monthly_sales = df['Monthly Sales']
fitted_values = results.fittedvalues

mse = mean_squared_error(monthly_sales, fitted_values)
rmse = np.sqrt(mse)
mae = mean_absolute_error(monthly_sales, fitted_values)

print("\nModel Performance Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
# Step 6: Forecast for the next 4 periods
forecast_periods = 4
forecast = results.get_forecast(steps=forecast_periods)
mean_forecast = forecast.predicted_mean

# Get confidence intervals for the forecast
conf_int_95 = forecast.conf_int(alpha=0.05)
conf_int_80 = forecast.conf_int(alpha=0.20)
conf_int_70 = forecast.conf_int(alpha=0.30)

# Extract the final forecast values and the confidence intervals
last_forecast = mean_forecast.iloc[-1]
ranges_95_lower = conf_int_95.iloc[-1, 0]
ranges_95_upper = conf_int_95.iloc[-1, 1]
ranges_80_lower = conf_int_80.iloc[-1, 0]
ranges_80_upper = conf_int_80.iloc[-1, 1]
ranges_70_lower = conf_int_70.iloc[-1, 0]
ranges_70_upper = conf_int_70.iloc[-1, 1]
```

```python
# Step 7: Display forecast information
print("\nForecast Ranges for Final Period:")
print(f"Point Forecast: ${last_forecast:,.2f}")
print("\nConfidence Intervals:")
print(f"95% CI: ${ranges_95_lower:,.2f} to ${ranges_95_upper:,.2f}")
print(f"80% CI: ${ranges_80_lower:,.2f} to ${ranges_80_upper:,.2f}")
print(f"70% CI: ${ranges_70_lower:,.2f} to ${ranges_70_upper:,.2f}")
```

```python
# Step 8: Calculate and display interval widths as percentage of forecast
print("\nInterval Widths as Percentage of Forecast:")
print(f"95% CI: ±{((ranges_95_upper - ranges_95_lower)/2/last_forecast*100):,.1f}%")
print(f"80% CI: ±{((ranges_80_upper - ranges_80_lower)/2/last_forecast*100):,.1f}%")
print(f"70% CI: ±{((ranges_70_upper - ranges_70_lower)/2/last_forecast*100):,.1f}%")
```

```python
# Step 9: Create DataFrame for forecast details
forecast_df = pd.DataFrame({
    'Forecast': mean_forecast,
    '95% Lower': conf_int_95.iloc[:, 0],
    '95% Upper': conf_int_95.iloc[:, 1],
    '80% Lower': conf_int_80.iloc[:, 0],
    '80% Upper': conf_int_80.iloc[:, 1],
    '70% Lower': conf_int_70.iloc[:, 0],
    '70% Upper': conf_int_70.iloc[:, 1]
})                          

print("\nDetailed Forecast with Confidence Intervals:")
forecast_df
```

```python
# Step 10: Plot the forecast and confidence intervals
plt.figure(figsize=(10, 6))
plt.plot(df['Monthly Sales'], label='Original Data', color='blue')
plt.plot(pd.date_range(df.index[-1], periods=4, freq='M'), mean_forecast, label='Forecast', color='red')
```

```python
# Add shaded areas for confidence intervals
plt.fill_between(pd.date_range(df.index[-1], periods=4, freq='M'), 
                 conf_int_95.iloc[:, 0], conf_int_95.iloc[:, 1], color='red', alpha=0.1, label='95% CI')

plt.fill_between(pd.date_range(df.index[-1], periods=4, freq='M'), 
                 conf_int_80.iloc[:, 0], conf_int_80.iloc[:, 1], color='orange', alpha=0.2, label='80% CI')

plt.fill_between(pd.date_range(df.index[-1], periods=4, freq='M'), 
                 conf_int_70.iloc[:, 0], conf_int_70.iloc[:, 1], color='green', alpha=0.3, label='70% CI')

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('ARIMA Forecast for Monthly Sales with Confidence Intervals')
plt.legend()
plt.xticks(rotation=45)

# Add a summary info box to the plot
info_text = (f'Final Forecast: ${last_forecast:,.2f}\n\n'
             f'95% CI: ${ranges_95_lower:,.2f} to ${ranges_95_upper:,.2f}\n'
             f'80% CI: ${ranges_80_lower:,.2f} to ${ranges_80_upper:,.2f}\n'
             f'70% CI: ${ranges_70_lower:,.2f} to ${ranges_70_upper:,.2f}\n\n'
             f'MAE: {mae:.2f}\nMSE: {mse:.2f}\nRMSE: {rmse:.2f}')

plt.text(0.02, 0.98, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.show()
```

## Holt-Winters Model

```python
# Step 2: Train Holt-Winters Exponential Smoothing model
# Seasonal='add' or 'mul' depends on the pattern of seasonality in your data
hw_model = ExponentialSmoothing(
    df['Monthly Sales'], 
    seasonal='add',  # or 'mul' if you have multiplicative seasonality
    seasonal_periods=12,  # Monthly seasonality
    trend='add',  # Additive trend
    damped_trend=True  # Optional, dampens the trend over time
).fit()

# Print model summary
print("\nHolt-Winters Model Summary:")
print(hw_model.summary())
```

```python
# Step 3: Calculate model performance metrics on fitted values
fitted_values = hw_model.fittedvalues
monthly_sales = df['Monthly Sales']

mse = mean_squared_error(monthly_sales, fitted_values)
rmse = np.sqrt(mse)
mae = mean_absolute_error(monthly_sales, fitted_values)

print("\nModel Performance Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

```python
# Step 4: Forecast for the next 4 periods
forecast_periods = 4
forecast = hw_model.forecast(steps=forecast_periods)
```

```python
# Step 5: Generate confidence intervals for the forecast
# Approximate the intervals using standard deviation of residuals (for simplicity)
residuals = monthly_sales - fitted_values
std_error = np.std(residuals)

conf_int_95 = pd.DataFrame({
    'Lower': forecast - 1.96 * std_error,
    'Upper': forecast + 1.96 * std_error
})

conf_int_80 = pd.DataFrame({
    'Lower': forecast - 1.28 * std_error,
    'Upper': forecast + 1.28 * std_error
})

conf_int_70 = pd.DataFrame({
    'Lower': forecast - 1.04 * std_error,
    'Upper': forecast + 1.04 * std_error
})

# Extract the final forecast values and the confidence intervals
last_forecast = forecast.iloc[-1]
ranges_95_lower = conf_int_95.iloc[-1, 0]
ranges_95_upper = conf_int_95.iloc[-1, 1]
ranges_80_lower = conf_int_80.iloc[-1, 0]
ranges_80_upper = conf_int_80.iloc[-1, 1]
ranges_70_lower = conf_int_70.iloc[-1, 0]
ranges_70_upper = conf_int_70.iloc[-1, 1]
```

```python
# Step 8: Plot the forecast and confidence intervals
plt.figure(figsize=(10, 6))
plt.plot(df['Monthly Sales'], label='Original Data', color='blue')
plt.plot(pd.date_range(df.index[-1], periods=forecast_periods, freq='M'), forecast, label='Forecast', color='red')

# Add shaded areas for confidence intervals
plt.fill_between(pd.date_range(df.index[-1], periods=forecast_periods, freq='M'), 
                 conf_int_95['Lower'], conf_int_95['Upper'], color='red', alpha=0.1, label='95% CI')

plt.fill_between(pd.date_range(df.index[-1], periods=forecast_periods, freq='M'), 
                 conf_int_80['Lower'], conf_int_80['Upper'], color='orange', alpha=0.2, label='80% CI')

plt.fill_between(pd.date_range(df.index[-1], periods=forecast_periods, freq='M'), 
                 conf_int_70['Lower'], conf_int_70['Upper'], color='green', alpha=0.3, label='70% CI')

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('SES Forecast for Monthly Sales with Confidence Intervals')
plt.legend()

# Add a summary info box to the plot
info_text = (f'Final Forecast: ${last_forecast:,.2f}\n\n'
             f'95% CI: ${ranges_95_lower:,.2f} to ${ranges_95_upper:,.2f}\n'
             f'80% CI: ${ranges_80_lower:,.2f} to ${ranges_80_upper:,.2f}\n'
             f'70% CI: ${ranges_70_lower:,.2f} to ${ranges_70_upper:,.2f}\n\n'
             f'MAE: {mae:.2f}\nMSE: {mse:.2f}\nRMSE: {rmse:.2f}')

plt.text(0.02, 0.98, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.show()
```

To compute the MAE (Mean Absolute Error), MSE (Mean Squared Error), and RMSE (Root Mean Squared Error) for each of the models (ARIMA, SES, Holt-Winters, and ETS), I will need to calculate them using the residuals (difference between actual and predicted values) for each model. Here's the process I'll follow for each model:



```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import Holt
from statsmodels.tsa.api import ETSModel
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Assuming 'df' is your DataFrame and 'Monthly Sales' is the target variable
# df = pd.read_csv('clean_monthly_sales_data.csv')
monthly_sales = df['Monthly Sales']
```

```python
# Step 1: ARIMA Model
auto_arima_order = (2, 0, 2)  # Example order, you should update it based on your auto_arima results
auto_seasonal_order = (1, 0, 2, 12)  # Example seasonal order, update as needed

arima_model = ARIMA(monthly_sales, order=auto_arima_order, seasonal_order=auto_seasonal_order)
arima_results = arima_model.fit()

arima_forecast = arima_results.fittedvalues
arima_mae = mean_absolute_error(monthly_sales, arima_forecast)
arima_mse = mean_squared_error(monthly_sales, arima_forecast)
arima_rmse = np.sqrt(arima_mse)
```

```python
# Step 2: Simple Exponential Smoothing (SES) Model
ses_model = SimpleExpSmoothing(monthly_sales).fit()
ses_forecast = ses_model.fittedvalues
ses_mae = mean_absolute_error(monthly_sales, ses_forecast)
ses_mse = mean_squared_error(monthly_sales, ses_forecast)
ses_rmse = np.sqrt(ses_mse)
```

```python
# Step 3: Holt-Winters Model (Triple Exponential Smoothing)
holt_winters_model = ExponentialSmoothing(monthly_sales, seasonal='add', seasonal_periods=12).fit()
holt_winters_forecast = holt_winters_model.fittedvalues
holt_winters_mae = mean_absolute_error(monthly_sales, holt_winters_forecast)
holt_winters_mse = mean_squared_error(monthly_sales, holt_winters_forecast)
holt_winters_rmse = np.sqrt(holt_winters_mse)
```

```python

# Step 4: ETS Model
ets_model = ETSModel(monthly_sales, error="add", trend="add", seasonal="add", seasonal_periods=12).fit()
ets_forecast = ets_model.fittedvalues
ets_mae = mean_absolute_error(monthly_sales, ets_forecast)
ets_mse = mean_squared_error(monthly_sales, ets_forecast)
ets_rmse = np.sqrt(ets_mse)
```

```python
# Step 5: Create a DataFrame to display results in tabular form
performance_data = {
    'Model': ['ARIMA', 'SES', 'Holt-Winters', 'ETS'],
    'MAE': [arima_mae, ses_mae, holt_winters_mae, ets_mae],
    'MSE': [arima_mse, ses_mse, holt_winters_mse, ets_mse],
    'RMSE': [arima_rmse, ses_rmse, holt_winters_rmse, ets_rmse]
}

performance_df = pd.DataFrame(performance_data)

# Step 6: Display the performance metrics as a table
print("\nModel Performance Metrics:")
print(performance_df)

performance_df
```

```python
# Optional: Visualize the performance metrics using a bar plot
performance_df.set_index('Model')[['MAE', 'MSE', 'RMSE']].plot(kind='bar', figsize=(10, 6))
plt.title('Model Performance Comparison (ARIMA, SES, Holt-Winters, ETS)')
plt.ylabel('Error Value')
plt.xticks(rotation=0)
plt.show()
```

## Analysis of the Metrics
#### 1. MAE (Mean Absolute Error)

 - Measures the average absolute difference between observed and predicted values.
 - Holt-Winters (24,575.71) performs the best, having the lowest MAE.
 - ARIMA (25,151.26) comes second, with a slightly higher error.
 - SES (27,700.01) and ETS (27,855.77) have the highest MAE, indicating they are less accurate.

#### 2. MSE (Mean Squared Error)

 - Squares the differences, giving more weight to larger errors.
 - Holt-Winters (1,004,984,000) performs best, as it has the lowest MSE.
 - ARIMA (1,007,609,000) is very close to Holt-Winters, indicating similar performance.
 - SES (1,167,818,000) and ETS (1,247,107,000) have much larger MSE values, indicating larger prediction errors.
#### RMSE (Root Mean Squared Error)

 - Measures the square root of MSE, in the same units as the original data.
 - Holt-Winters (31,701.48) performs best with the smallest RMSE.
 - ARIMA (31,742.85) is very close to Holt-Winters, suggesting ARIMA also performs well.
 - SES (34,173.36) and ETS (35,314.40) have larger RMSE, indicating poorer performance.
